#1. Análisis de facturación



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Análisis de facturación para cada tienda

total_tienda1 = tienda['Precio'].sum()
promedio_precio_tienda1 = tienda['Precio'].mean()
cantidad_ventas_tienda1 = tienda.shape[0] # Número de filas como aproximación a la cantidad de ventas

total_tienda2 = tienda2['Precio'].sum()
promedio_precio_tienda2 = tienda2['Precio'].mean()
cantidad_ventas_tienda2 = tienda2.shape[0]

total_tienda3 = tienda3['Precio'].sum()
promedio_precio_tienda3 = tienda3['Precio'].mean()
cantidad_ventas_tienda3 = tienda3.shape[0]

total_tienda4 = tienda4['Precio'].sum()
promedio_precio_tienda4 = tienda4['Precio'].mean()
cantidad_ventas_tienda4 = tienda4.shape[0]


print("--- Análisis de Facturación por Tienda ---")
print(f"Tienda 1:")
print(f"  Total Facturado: ${total_tienda1:,.2f}")
print(f"  Precio Promedio por Venta: ${promedio_precio_tienda1:,.2f}")
print(f"  Cantidad de Ventas (estimado): {cantidad_ventas_tienda1}")
print("-" * 20)

print(f"Tienda 2:")
print(f"  Total Facturado: ${total_tienda2:,.2f}")
print(f"  Precio Promedio por Venta: ${promedio_precio_tienda2:,.2f}")
print(f"  Cantidad de Ventas (estimado): {cantidad_ventas_tienda2}")
print("-" * 20)

print(f"Tienda 3:")
print(f"  Total Facturado: ${total_tienda3:,.2f}")
print(f"  Precio Promedio por Venta: ${promedio_precio_tienda3:,.2f}")
print(f"  Cantidad de Ventas (estimado): {cantidad_ventas_tienda3}")
print("-" * 20)

print(f"Tienda 4:")
print(f"  Total Facturado: ${total_tienda4:,.2f}")
print(f"  Precio Promedio por Venta: ${promedio_precio_tienda4:,.2f}")
print(f"  Cantidad de Ventas (estimado): {cantidad_ventas_tienda4}")
print("-" * 20)
```

```python
import pandas as pd
import altair as alt

data = {'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
        'Total Facturado': [1150880400.00, 1116343500.00, 1098019600.00, 1038375700.00],
        'Precio Promedio por Venta': [487867.91, 473227.43, 465459.77, 440362.89]}

df_billing = pd.DataFrame(data)

chart_total = alt.Chart(df_billing).mark_bar().encode(
    x=alt.X('Tienda', title='Tienda'),
    y=alt.Y('Total Facturado', title='Total Facturado'),
    tooltip=['Tienda', 'Total Facturado']
).properties(
    title='Total Facturado por Tienda'
)

chart_average = alt.Chart(df_billing).mark_bar().encode(
    x=alt.X('Tienda', title='Tienda'),
    y=alt.Y('Precio Promedio por Venta', title='Precio Promedio por Venta'),
    tooltip=['Tienda', 'Precio Promedio por Venta']
).properties(
    title='Precio Promedio por Venta por Tienda'
)

chart = chart_total | chart_average
chart.display()
```

# 2. Ventas por categoría

```python
# Análisis de ventas por categoría para cada tienda

ventas_por_categoria1 = tienda.groupby('Categoría del Producto')['Precio'].sum().reset_index()
ventas_por_categoria1 = ventas_por_categoria1.rename(columns={'Precio': 'Total Facturado'})
ventas_por_categoria1['Tienda'] = 'Tienda 1'


ventas_por_categoria2 = tienda2.groupby('Categoría del Producto')['Precio'].sum().reset_index()
ventas_por_categoria2 = ventas_por_categoria2.rename(columns={'Precio': 'Total Facturado'})
ventas_por_categoria2['Tienda'] = 'Tienda 2'

ventas_por_categoria3 = tienda3.groupby('Categoría del Producto')['Precio'].sum().reset_index()
ventas_por_categoria3 = ventas_por_categoria3.rename(columns={'Precio': 'Total Facturado'})
ventas_por_categoria3['Tienda'] = 'Tienda 3'

ventas_por_categoria4 = tienda4.groupby('Categoría del Producto')['Precio'].sum().reset_index()
ventas_por_categoria4 = ventas_por_categoria4.rename(columns={'Precio': 'Total Facturado'})
ventas_por_categoria4['Tienda'] = 'Tienda 4'

ventas_por_categoria_total = pd.concat([ventas_por_categoria1, ventas_por_categoria2, ventas_por_categoria3, ventas_por_categoria4])

# Formatear la columna 'Total Facturado'
ventas_por_categoria_total['Total Facturado'] = ventas_por_categoria_total['Total Facturado'].apply(lambda x: f"${x:,.2f}")

print("--- Reporte de Ventas por Categoría por Tienda ---")
display(ventas_por_categoria_total)
```

```python
import altair as alt

# Convertir 'Total Facturado' de nuevo a numérico para graficar
ventas_por_categoria_total['Total Facturado'] = ventas_por_categoria_total['Total Facturado'].replace({r'\$': '', ',': ''}, regex=True).astype(float)

chart_ventas_categoria = alt.Chart(ventas_por_categoria_total).mark_bar().encode(
    x=alt.X('Tienda', title='Tienda'),
    y=alt.Y('Total Facturado', title='Total Facturado'),
    color='Categoría del Producto',  # Apilar por categoría
    tooltip=['Tienda', 'Categoría del Producto', alt.Tooltip('Total Facturado', format='$,.2f')]
).properties(
    title='Ventas por Categoría por Tienda (Apilado)'
)

chart_ventas_categoria.display()
```

# 4. Productos más y menos vendidos

```python
# Análisis de productos más y menos vendidos por tienda

print("--- Productos Más y Menos Vendidos por Tienda ---")

# Tienda 1
productos_vendidos1 = tienda['Producto'].value_counts()
top_5_vendidos1 = productos_vendidos1.head(5)
bottom_5_vendidos1 = productos_vendidos1.tail(5)

print("\nTienda 1:")
print("Top 5 Productos Más Vendidos:")
for producto, count in top_5_vendidos1.items():
    print(f"- {producto}: {count}")
print("\nTop 5 Productos Menos Vendidos:")
for producto, count in bottom_5_vendidos1.items():
    print(f"- {producto}: {count}")
print("-" * 20)

# Tienda 2
productos_vendidos2 = tienda2['Producto'].value_counts()
top_5_vendidos2 = productos_vendidos2.head(5)
bottom_5_vendidos2 = productos_vendidos2.tail(5)

print("\nTienda 2:")
print("Top 5 Productos Más Vendidos:")
for producto, count in top_5_vendidos2.items():
    print(f"- {producto}: {count}")
print("\nTop 5 Productos Menos Vendidos:")
for producto, count in bottom_5_vendidos2.items():
    print(f"- {producto}: {count}")
print("-" * 20)

# Tienda 3
productos_vendidos3 = tienda3['Producto'].value_counts()
top_5_vendidos3 = productos_vendidos3.head(5)
bottom_5_vendidos3 = productos_vendidos3.tail(5)

print("\nTienda 3:")
print("Top 5 Productos Más Vendidos:")
for producto, count in top_5_vendidos3.items():
    print(f"- {producto}: {count}")
print("\nTop 5 Productos Menos Vendidos:")
for producto, count in bottom_5_vendidos3.items():
    print(f"- {producto}: {count}")
print("-" * 20)

# Tienda 4
productos_vendidos4 = tienda4['Producto'].value_counts()
top_5_vendidos4 = productos_vendidos4.head(5)
bottom_5_vendidos4 = productos_vendidos4.tail(5)

print("\nTienda 4:")
print("Top 5 Productos Más Vendidos:")
for producto, count in top_5_vendidos4.items():
    print(f"- {producto}: {count}")
print("\nTop 5 Productos Menos Vendidos:")
for producto, count in bottom_5_vendidos4.items():
    print(f"- {producto}: {count}")
print("-" * 20)
```

```python
import pandas as pd
import altair as alt

# Crear DataFrames para los productos más y menos vendidos
data_top_bottom = {
    'Tienda': [],
    'Producto': [],
    'Cantidad': [],
    'Tipo': []
}

for tienda_name, top_products, bottom_products in [
    ('Tienda 1', top_5_vendidos1, bottom_5_vendidos1),
    ('Tienda 2', top_5_vendidos2, bottom_5_vendidos2),
    ('Tienda 3', top_5_vendidos3, bottom_5_vendidos3),
    ('Tienda 4', top_5_vendidos4, bottom_5_vendidos4)
]:
    for producto, count in top_products.items():
        data_top_bottom['Tienda'].append(tienda_name)
        data_top_bottom['Producto'].append(producto)
        data_top_bottom['Cantidad'].append(count)
        data_top_bottom['Tipo'].append('Más Vendido')

    for producto, count in bottom_products.items():
        data_top_bottom['Tienda'].append(tienda_name)
        data_top_bottom['Producto'].append(producto)
        data_top_bottom['Cantidad'].append(count)
        data_top_bottom['Tipo'].append('Menos Vendido')

df_top_bottom = pd.DataFrame(data_top_bottom)

# Crear gráficos circulares para los productos más y menos vendidos por tienda
base = alt.Chart(df_top_bottom).encode(
    theta=alt.Theta("Cantidad", stack=True)
).properties(
    width=200,
    height=200
)

pie = base.mark_arc(outerRadius=80).encode(
    color=alt.Color("Producto", legend=None), # Eliminar leyenda para el color
    order=alt.Order("Cantidad", sort="descending"),
    tooltip=["Producto", "Cantidad", alt.Tooltip("Cantidad", format=".1%")] # Formato de tooltip como porcentaje
)

text_quantity = base.mark_text(radius=60, fill="black").encode( # Texto dentro del arco
    text=alt.Text("Cantidad"),
    order=alt.Order("Cantidad", sort="descending")
)

text_product = base.mark_text(radius=100, fill="black").encode( # Texto fuera del arco
    text=alt.Text("Producto"),
    order=alt.Order("Cantidad", sort="descending")
)


chart_pie = (pie + text_quantity + text_product).facet(
    column=alt.Column('Tienda:N', header=alt.Header(title='Tienda', titleOrient="bottom", labelOrient="bottom")), # Título añadido a la faceta de columna
    row=alt.Row('Tipo:N', header=alt.Header(title='Tipo de Producto', titleOrient="right", labelOrient="right")) # Título añadido a la faceta de fila
).properties(
    title='Productos Más y Menos Vendidos por Tienda (Gráficos Circulares)'
)

chart_pie.display()
```
